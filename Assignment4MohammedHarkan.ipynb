```python
# Python and Pandas Assessment Notebook

import pandas as pd
import numpy as np
from functools import reduce

# Task 1: Create a DataFrame using Pandas
# Create a DataFrame with the following data:
# Name: Alice, Bob, Charlie, David
# Age: 25, 30, 35, 28
# City: New York, San Francisco, Los Angeles, Chicago

# Your code here:
# Create a dictionary that contains the data for the DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Age": [25, 30, 35, 28],
    "City": ["New York", "San Francisco", "Los Angeles", "Chicago"]
}

# Convert the dictionary into a Pandas DataFrame
df = pd.DataFrame(data)

# Print the DataFrame to show the result
print("Task 1: Original DataFrame")
print(df)

# Print the data types to make sure each column has the correct type
print("\nData types:")
print(df.dtypes)


# Task 2: Row and Column Manipulation
# Drop the 'City' column from the DataFrame created in Task 1

# Your code here:
# Create a new DataFrame without the City column
df_without_city = df.drop("City", axis=1)

# Print the new DataFrame
print("Task 2: DataFrame after dropping City column")
print(df_without_city)

# Verify the remaining columns
print("\nRemaining columns:")
print(df_without_city.columns)


# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them

# Your code here:
# Create a new DataFrame with some null values
df_null = pd.DataFrame({
    "Name": ["Alice", None, "Charlie", "David"],
    "Age": [25, 30, None, 28],
    "City": ["New York", "San Francisco", None, "Chicago"]
})

# Print the DataFrame before filling null values
print("Task 3: DataFrame with null values")
print(df_null)

# Check the number of null values in each column
print("\nNull values in each column:")
print(df_null.isnull().sum())

# Fill null values with appropriate values
df_filled = df_null.fillna({
    "Name": "Unknown",
    "Age": 0,
    "City": "Unknown"
})

# Print the DataFrame after filling null values
print("\nDataFrame after filling null values")
print(df_filled)

# Task 4: GroupBy and Describe
# Using the following DataFrame, group by 'Category' and describe the 'Value' column

df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 15, 25, 30, 35]
})

# Your code here:
# Group the data by Category and calculate statistics for Value
grouped_stats = df_group.groupby("Category")["Value"].describe()

# Print the statistics
print("Task 4: GroupBy and Describe")
print(grouped_stats)


# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

# Your code here:

# Concatenate df1 and df2 vertically
df_concat = pd.concat([df1, df2], ignore_index=True)

# Print the concatenated DataFrame
print("Task 5: Concatenated DataFrame")
print(df_concat)

# Merge df1 and df3 horizontally
df_merged = pd.concat([df1, df3], axis=1)

# Print the merged DataFrame
print("\nMerged DataFrame")
print(df_merged)


# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Your code here:
# Create a tuple of fruits
fruits_tuple = ("Apple", "Banana", "Orange")

# Create a set of numbers
numbers_set = {1, 2, 3, 4, 5}

# Print original values
print("Original Tuple:")
print(fruits_tuple)

print("\nOriginal Set:")
print(numbers_set)

# Try to add an element to the tuple
try:
    fruits_tuple.append("Mango")
except AttributeError as error:
    print("\nCannot add to tuple:")
    print(error)

# Add an element to the set
numbers_set.add(6)

print("\nSet after adding 6:")
print(numbers_set)


# Task 7: Dictionaries
# Create a dictionary of student names and their scores.
# Then, update a student's score and add a new student.

# Your code here:

# Create a dictionary of students and their scores
students = {
    "Ali": 90,
    "Sara": 85,
    "Omar": 95
}

# Print original dictionary
print("Original Dictionary:")
print(students)

# Update Sara's score
students["Sara"] = 88

# Add a new student
students["Mona"] = 92

# Print updated dictionary
print("\nUpdated Dictionary:")
print(students)

# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

# Your code here:
# Regular function to calculate the square of a number
def square(number):
    return number ** 2

# Lambda function to calculate the square of a number
square_lambda = lambda number: number ** 2

# Test both functions
print("Task 8: Functions and Lambda")

print("Regular Function:")
print(square(4))
print(square(7))

print("\nLambda Function:")
print(square_lambda(4))
print(square_lambda(7))


# Task 9: Iterators and Generators
# Create an iterator for the first 5 even numbers.
# Then, create a generator for the same.

# Your code here:

# Iterator class for first 5 even numbers
class EvenIterator:
    def __init__(self):
        self.current = 2
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < 5:
            value = self.current
            self.current += 2
            self.count += 1
            return value
        else:
            raise StopIteration

# Generator function for first 5 even numbers
def even_generator():
    for number in range(2, 12, 2):
        yield number

print("Task 9: Iterators and Generators")

print("Iterator Output:")
for num in EvenIterator():
    print(num)

print("\nGenerator Output:")
for num in even_generator():
    print(num)


# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Your code here:
# Use map to square all numbers
squared_numbers = list(map(lambda x: x ** 2, numbers))

# Use reduce to find the product of all numbers
product = reduce(lambda x, y: x * y, numbers)

# Use filter to get even numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("Task 10: Map, Reduce, and Filter")

print("Squared Numbers:")
print(squared_numbers)

print("\nProduct of Numbers:")
print(product)

print("\nEven Numbers:")
print(even_numbers)


# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

# Your code here:
# Create Rectangle class
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

# Create rectangle objects
rectangle1 = Rectangle(5, 3)
rectangle2 = Rectangle(8, 4)

print("Task 11: Rectangle Class")

print("Rectangle 1")
print("Area:", rectangle1.area())
print("Perimeter:", rectangle1.perimeter())

print("\nRectangle 2")
print("Area:", rectangle2.area())
print("Perimeter:", rectangle2.perimeter())

# Task 12: Pandas Data Analysis
# Using the following DataFrame, perform these tasks:
# 1. Find the average salary by department
# 2. Get the names of employees with salary > 60000
# 3. Add a new column 'Bonus' which is 10% of salary

df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# Your code here:

# Calculate average salary by department
average_salary = df_employees.groupby("Department")["Salary"].mean()

# Get employees with salary greater than 60000
high_salary_employees = df_employees[df_employees["Salary"] > 60000]["Name"]

# Add Bonus column
df_employees["Bonus"] = df_employees["Salary"] * 0.10

print("Task 12: Pandas Data Analysis")

print("\nAverage Salary by Department:")
print(average_salary)

print("\nEmployees with Salary > 60000:")
print(high_salary_employees)

print("\nDataFrame with Bonus Column:")
print(df_employees)
```
